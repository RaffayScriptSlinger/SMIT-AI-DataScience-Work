```python
def are_anagrams(str1, str2):
    s1 = str1.replace(" ", "").lower()
    s2 = str2.replace(" ", "").lower()

    return sorted(s1) == sorted(s2)
    
print(are_anagrams("listen", "silent"))        
print(are_anagrams("Debit Card", "Bad Credit"))
print(are_anagrams("rat", "tart"))  
```
