### Constructor Or Innitializer

```python
class Mobile() :  
    def __init__(self,model,Description):
        
    # Ya Attribute Han
        self.model = model
        self.Description = Description

    #Ya Wala Methods Han
    def On_Moile(self):
        print("Mobile Phone Is One",self.model)
    def off_Mobile():
        print("Mobile Phone Is Off")
```

```python
Mobile_1 = Mobile( "Iphone","ALL GOOD ")
```

```python
print(Mobile_1.model)
```

```python
print(Mobile_1.Description)
```
