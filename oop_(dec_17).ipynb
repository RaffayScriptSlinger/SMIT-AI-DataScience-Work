```python
class Laptop():
    def __init__(self ,brand, color, ram, SSD, processer, genration):
        self.__brand = brand
        self.__color = color
        self.__ram = ram
        self.__SSD = SSD
        self._processer = processer
        self.genration = genration

    def start(self):
        print("laptop power on ho gaya")
    def shutdown(self):
        print("laptop shutdown ho gaya")
    def hello(self):
        print(f"mera pass laptop {self.brand} conpany ka he")
```

```python
lap1 = Laptop("hp", "black", "20gb", "512gb", "corei7", "10th")
```

```python
lap1.brand
```

```python
lap1.processer
```

```python
lap1.SSD
```

```python
lap1.hello()
```

```python
lap1.shutdown()
```

# pillar of OOP
     - Encapsulation
     - Inheritance
     - Polymorphism
     - Abstraction

```python
class Std_data():
    def __init__(self, name, father_name, age, roll_number, gender, Class):
        self.__name = name
        self.__father_name = father_name
        self.__age = age
        self.__roll_number = roll_number
        self.__gender = gender
        self.__Class = Class

    def std_info(self):
        print(f"student name is: {self.__name}")
        print(f"student father name is: {self.__father_name}")
        print(f"student age is: {self.__age}")
        print(f"student roll_number is: {self.__roll_number}")
        print(f"student gender is: {self.__gender}")
        print(f"student Class is: {self.__Class}")
```

```python
std1 = Std_data("raza", "anwer", "21", "1001", "male", "12th")
```

```python
std1.std_info()
```

```python
std1.Class
```

```python
std1.std_info()
```
