```python
class battery:
    def __init__(self, plates, Type, Duration, RechargeAble):
        self.plates = plates
        self.Type = Type
        self.Duration = Duration
        self.RechargeAble = RechargeAble

    def charging(self):
        if self.RechargeAble:
            print("Yes it is Chargable")
        else:
            print("NO Is Not")
```
